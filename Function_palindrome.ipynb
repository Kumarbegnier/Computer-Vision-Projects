```python
def is_palindrome_two_pointer(value):
    s = str(value)  # Convert to string
    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False  # Not a palindrome
        left += 1
        right -= 1

    return True  # All characters matched

# Test the function
print(is_palindrome_two_pointer("madam"))  # True
print(is_palindrome_two_pointer("hello"))  # False
print(is_palindrome_two_pointer(121))      # True
print(is_palindrome_two_pointer(123))      # False
```
